```python
import matplotlib.pyplot as plt

# Données
leagues = ['Premier League', 'Ligue 1', 'La Liga']
draw_percentages = [21.58, 26.47, 22.01]
win_percentages = [39.21, 36.76, 61.00]

# Configuration des barres
x = range(len(leagues))
width = 0.35

fig, ax = plt.subplots()

# Barres pour les pourcentages de matchs nuls
draw_bars = ax.bar(x, draw_percentages, width, label='Matchs nuls')

# Barres pour les pourcentages de matchs gagnés
win_bars = ax.bar([p + width for p in x], win_percentages, width, label='Matchs gagnés')

# Labels et titre
ax.set_xlabel('Ligues')
ax.set_ylabel('Pourcentage (%)')
ax.set_title('Comparaison des pourcentages de matchs nuls et de matchs gagnés')
ax.set_xticks([p + width/2 for p in x])
ax.set_xticklabels(leagues)
ax.legend()

# Affichage des pourcentages sur les barres
for bar in draw_bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

for bar in win_bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

# Affichage du graphique
plt.show()
```
